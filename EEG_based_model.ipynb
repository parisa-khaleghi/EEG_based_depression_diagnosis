```python
import pandas as pd

# Define the path in your Google Drive
drive_path = '/content/drive/My Drive/all_eeg_features.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(drive_path)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
# Create an array of DataFrames
num_patients = df['record_id'].nunique()
patient_dataframes = []

for i in range(num_patients):
    patient_df = df[df['record_id'] == i].reset_index(drop=True)
    patient_dataframes.append(patient_df)

# Example: Access the DataFrame for patient with record_id 0
patient_dataframes[0].head()
```

```python
# Concatenate list of DataFrames into a single DataFrame if df is a list

df = patient_dataframes
if isinstance(df, list):
    df = pd.concat(df, ignore_index=True)

# List of columns to take min, max, and mean
min_columns = ['Wavelet Detailed Entropy', 'Mobility', 'Complexity']
max_columns = ['FFT Theta Max Power', 'Delta/Alpha']
mean_columns = [col for col in df.columns if col not in min_columns + max_columns + ['record_id'] + ['depressed']]

# Initialize an empty list to store the aggregated data
aggregated_data = []

# Iterate over each unique record_id
for record_id in df['record_id'].unique():
    patient_data = df[df['record_id'] == record_id]
    aggregated_row = {
        'record_id': record_id,
    }
    # Calculate min, max, and mean for the respective columns
    for col in min_columns:
        aggregated_row[col] = patient_data[col].min()
    for col in max_columns:
        aggregated_row[col] = patient_data[col].max()
    for col in mean_columns:
        aggregated_row[col] = patient_data[col].mean()
    aggregated_data.append(aggregated_row)

# Convert the aggregated data to a DataFrame
aggregated_df = pd.DataFrame(aggregated_data)

# Print the aggregated DataFrame
print(aggregated_df)
```

```python
# Define normal ranges
normal_ranges = {
    'FFT Theta Max Power': (0.0005, 0.0007),
    'Wavelet Detailed Entropy': (0.0012, 0.0015),
    'Mobility': (0.30, 0.40),
    'Complexity': (4.0, 5.0),
    'Delta/Alpha': (50, 70)
}

def is_non_depressed(row):
    for feature, (low, high) in normal_ranges.items():
        if not (low <= row[feature] <= high):
            return 0
    return 1

# Add a new column to indicate depression status
aggregated_df['depressed'] = aggregated_df.apply(is_non_depressed, axis=1)

# Count the number of non-depressed and depressed patients
non_depressed_count = aggregated_df['depressed'].sum()
depressed_count = len(aggregated_df) - non_depressed_count

print(f"Non-depressed patients: {non_depressed_count}")
print(f"Depressed patients: {depressed_count}")
```

```python
# Function to extract wavelet features for all frequency bands
def extract_wavelet_features(data, wavelet='db4'):
    wavelet_features = []
    for record in data:
        coeffs = pywt.wavedec(record, wavelet, level=5)
        features = np.hstack([np.mean(c) for c in coeffs])
        wavelet_features.append(features)
    return np.array(wavelet_features)

# Function to extract power spectral density features
def extract_psd_features(data, fs=256):
    psd_features = []
    for record in data:
        f, Pxx = welch(record, fs=fs)
        psd_features.append(np.hstack([np.mean(Pxx[freq_range[0]:freq_range[1]]) for freq_range in [(0, 4), (4, 8), (8, 13), (13, 30), (30, 50)]]))  # Delta, Theta, Alpha, Beta, Gamma bands
    return np.array(psd_features)

# Function to extract time-frequency features using STFT
def extract_stft_features(data, fs=256, nperseg=128):
    stft_features = []
    for record in data:
        f, t, Zxx = stft(record, fs=fs, nperseg=nperseg)
        features = np.abs(Zxx).mean(axis=1)  # Mean magnitude of STFT coefficients
        stft_features.append(features)
    return np.array(stft_features)

# Function to extract additional statistical features
def extract_statistical_features(data):
    statistical_features = []
    for record in data:
        features = [
            np.mean(record), np.std(record), skew(record), kurtosis(record)
        ]
        statistical_features.append(features)
    return np.array(statistical_features)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, Flatten, MaxPooling1D, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reshape data for CNN input
X = pca_result.reshape(pca_result.shape[0], pca_result.shape[1], 1)

# Print the shape of X to check
print("Shape of X:", X.shape)

# Ensure 'y' is a 1D array
y = aggregated_df['depressed'].values

# Print the shape of y to check
print("Shape of y:", y.shape)

# Continue with the rest of the code
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Reshape data for CNN input
X = pca_result.reshape(pca_result.shape[0], pca_result.shape[1], 1)

# Ensure 'y' is a 1D array
y = aggregated_df['depressed'].values

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Compute class weights to handle class imbalance
class_weights = {0: len(y) / (2 * np.bincount(y)[0]), 1: len(y) / (2 * np.bincount(y)[1])}

# Define the CNN model
model = Sequential()
model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(X.shape[1], 1)))
model.add(BatchNormalization())
model.add(MaxPooling1D(pool_size=2))
model.add(Conv1D(filters=128, kernel_size=3, activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.0001), loss='binary_crossentropy', metrics=['accuracy'])

# Define early stopping and learning rate reduction
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.00001)

# Train the model
history = model.fit(X_train, y_train, epochs=100, batch_size=64, validation_data=(X_test, y_test), callbacks=[early_stopping, reduce_lr], class_weight=class_weights)

# Evaluate the model
accuracy = model.evaluate(X_test, y_test)[1]
print(f"CNN Accuracy: {accuracy:.2f}")

# Predict and evaluate the model
y_pred = (model.predict(X_test) > 0.5).astype("int32")
print("Classification Report:\n", classification_report(y_test, y_pred))

# Visualize training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title('Loss over epochs')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='train_accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.title('Accuracy over epochs')
plt.legend()
plt.show()
```
